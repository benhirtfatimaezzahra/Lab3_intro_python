```python
print("=== Conversion de températures ===")
print("1. Celsius  -> Fahrenheit et Kelvin")
print("2. Fahrenheit -> Celsius et Kelvin")
choix =input("Votre choix: ")

if choix=="1":
    try:
        c = float(input("Entrez la température en celsius : "))
    except ValueError:
        print("Erreur")
        quit()
    f = c * 9/5 + 32
    k = c + 273.15
    print(f"{c}°C = {f}°F = {k}K")
elif choix == "2":
    try:
        f = float(input("Entrez la température en Fahrenheit  : "))
    except ValueError:
        print("Erreur ")
        quit()
    c = (f - 32) * 5/9
    k = c + 273.15

    print(f"{f}°F = {c}°C = {k}K")
else:
    print("Erreur : choix invalide.")
```
