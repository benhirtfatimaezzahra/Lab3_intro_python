```python
try:
    a = float(input("Entrez le premier nombre : "))
    b = float(input("Entrez le deuxième nombre : "))
except ValueError:
    print("Erreur : veuillez entrer des nombres valides.")
    quit()

print("\nChoisissez une opération :")
print(" +  pour l'addition")
print(" -  pour la soustraction")
print(" *  pour la multiplication")
print(" /  pour la division")

op = input("Votre choix: ")

if op == "+":
    result = a + b
elif op == "-":
    result = a - b
elif op == "*":
    result = a * b
elif op == "/":
    if b == 0:
        print("Erreur : division par zéro impossible.")
        quit()
    result = a / b
else:
    print("Erreur : opérateur invalide.")
    quit()

print(f"\n{a} {op} {b} = {result}")
```
